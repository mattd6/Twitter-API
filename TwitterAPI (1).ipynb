Imports, Twitter API Authentication

```python
import tweepy,json
import os
import pandas as pd
import sys
import csv

#Enter names of API keys and tokens. These act as login credentials
access_token="<insert access token>"
access_token_secret="<insert secret access token>"
consumer_key="<insert consumer key>"
consumer_secret="<insert secret consumer key>"

auth= tweepy.OAuthHandler(consumer_key,consumer_secret)
auth.set_access_token(access_token,access_token_secret)
api = tweepy.API(auth, wait_on_rate_limit=True)
```

Tell the program to search twitter for tweets containing the term(s) contained in search_words

```python
search_words = "are -filter:retweets"

date_since = "2014-03-12"
```

Collect tweets. .Cursor() searches for tweets containing the search word. Specify how many tweets to return by typing number in .items()

```python
tweets = tweepy.Cursor(api.search,
              q=search_words,
              lang="en",
              since=date_since).items(600)

# Iterate and print tweets
for tweet in tweets:
    print(tweet.text)
```

Store user tweets and users in a list

```python
tweets = tweepy.Cursor(api.search, 
                           q=search_words,
                           lang="en",
                           since=date_since).items(600)

users_locs = [[tweet.text, tweet.user.screen_name, tweet.user.id] for tweet in tweets]
users_locs
```

Put list into dataset

```python
df = pd.DataFrame(data=users_locs, 
                    columns=['text','user', "id"])
df
```
